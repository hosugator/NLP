### 3) TF-IDF (Term Frequency-Inverse Document Frequency)
- TfidfVectorizer는 텍스트 데이터에서 각 단어의 출현 빈도(TF, Term Frequency)와 역문서 빈도(IDF, Inverse Document Frequency)를 결합하여 벡터를 생성합니다. 단어가 문서 내에서 자주 등장하지만 전체 문서 집합에서는 드물게 등장하는 경우 더 높은 가중치를 부여합니다
- 단어의 출현 빈도(TF)와 역문서 빈도(IDF)를 결합하여 가중치를 계산합니다.
- 단어의 중요도를 반영하여 벡터를 생성합니다.
- 자주 등장하는 일반적인 단어는 낮은 가중치를, 드물게 등장하는 중요한 단어는 높은 가중치를 부여합니다.
- 사용 목적:
    - 단어의 중요도를 반영한 벡터화를 원할 때.
