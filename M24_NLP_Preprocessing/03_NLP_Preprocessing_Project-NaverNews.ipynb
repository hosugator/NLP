- 네이버 뉴스 - 정치 섹션 뉴스 수집하기

```python
# 네이버 뉴스 - 정치 섹션
import requests
from bs4 import BeautifulSoup
import json

def get_news_fixed(section_url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
    response = requests.get(section_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    articles = []
    seen_titles = set()
    
    # 여러 셀렉터 시도
    selectors = [
        'a[href*="article"]',      # article이 포함된 모든 링크
        'a[href*="mnews"]',        # mnews가 포함된 링크  
        'a.sa_text',               # 원래 셀렉터
        'a.nclicks\\(fls\\.list\\)'  # 네이버 클릭 추적 링크
    ]
    
    for selector in selectors:
        try:
            links = soup.select(selector)
            #print(f"{selector}: {len(links)}개 링크 발견")
            
            for link in links:
                title = link.get_text(strip=True)
                url = link.get('href', '')
                
                # URL 정규화
                if url.startswith('/'):
                    url = 'https://news.naver.com' + url
                elif url.startswith('//'):
                    url = 'https:' + url
                
                # 유효성 검사
                if (title and len(title) > 10 and len(title) < 200 and
                    title not in seen_titles and
                    ('article' in url or 'mnews' in url or 'news.naver.com' in url)):
                    
                    seen_titles.add(title)
                    articles.append({'title': title, 'url': url})
                    
        except Exception as e:
            print(f"{selector} 오류: {e}")
    
    return articles

# 테스트
news_100 = get_news_fixed('https://news.naver.com/section/100')
#print(f"수집된 기사 수: {len(news)}")
for i, article in enumerate(news_100):
    #print(f"{i+1}. {article['title']}")
    print(f"{article['title']}")
```

- 네이버 뉴스 - 정치, 경제 등 여러 섹션 뉴스 수집하기

```python
# 네이버 뉴스 - 전체 섹션 수집 및 저장 
import requests
from bs4 import BeautifulSoup
import json
import time

def get_news_fixed(section_url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
    response = requests.get(section_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    articles = []
    seen_titles = set()
    
    selectors = [
        'a[href*="article"]',
        'a[href*="mnews"]',  
        'a.sa_text',
        'a[onclick*="nclicks(fls.list)"]'
    ]
    
    for selector in selectors:
        try:
            links = soup.select(selector)
            
            for link in links:
                title = link.get_text(strip=True)
                url = link.get('href', '')
                
                if url.startswith('/'):
                    url = 'https://news.naver.com' + url
                elif url.startswith('//'):
                    url = 'https:' + url
                
                if (title and len(title) > 10 and len(title) < 200 and
                    title not in seen_titles and
                    ('article' in url or 'mnews' in url or 'news.naver.com' in url)):
                    
                    seen_titles.add(title)
                    articles.append({'title': title, 'url': url})
                    
        except Exception as e:
            pass
    
    return articles

def collect_all_news():
    sections = {
        '정치': 'https://news.naver.com/section/100',
        '경제': 'https://news.naver.com/section/101', 
        '사회': 'https://news.naver.com/section/102',
        '생활문화': 'https://news.naver.com/section/103',
        '세계': 'https://news.naver.com/section/104',
        'IT과학': 'https://news.naver.com/section/105',
        '연예': 'https://news.naver.com/section/106',
        '스포츠': 'https://news.naver.com/section/107'
    }
    
    all_articles = []
    
    for section_name, section_url in sections.items():
        articles = get_news_fixed(section_url)
        
        for article in articles:
            article['section'] = section_name
            all_articles.append(article)
            print(f"[{section_name}] {article['title']}")
        
        time.sleep(0.5)  # 0.5초 대기
    
    return all_articles

# 실행
all_articles = collect_all_news()

# JSON 저장
with open('all_news.json', 'w', encoding='utf-8') as f:
    json.dump(all_articles, f, ensure_ascii=False, indent=2)

print(f"\n총 {len(all_articles)}개 기사 수집 완료")
```

```python
import pandas as pd
import re


# 1. 간단한 클린징 함수
def simple_clean(text):
    """간단한 텍스트 클린징"""
    if not isinstance(text, str):
        return ""

    # HTML 태그 제거
    text = re.sub(r"<[^>]+>", "", text)  # code here

    # 특수문자 제거 (한글, 영문, 숫자, 공백만 보존)
    # \w = a-zA-Z가-힣0-9 (정규식)
    text = re.sub(r"[^a-zA-Z가-힣0-9\s]", " ", text)  # code here 

    # 연속된 공백을 하나로
    text = re.sub(r"\s+", " ", text)  # code here

    return text.strip()


# 2. 전체 데이터 처리

# df = pd.DataFrame(all_articles)
df = pd.read_csv("news_data.csv")
df["title_cleaned"] = df["title"].apply(simple_clean)

# 3. 모든 제목을 하나의 텍스트로 합치기
filtered_content = " ".join(df["title_cleaned"])
```

```python
#  예시 문장 : cleaning 순서 적용에 따른 결과 차이 확인 1
re_special = re.compile(r'[^a-zA-Z가-힣\s]')   # 한글/영문/공백만 보존 (숫자 제외됨)
re_html    = re.compile(r'<[^>]+>')
re_space   = re.compile(r'\s+')

def step_special(s: str) -> str:
    return re_special.sub(' ', s)

def step_html(s: str) -> str:
    return re_html.sub('', s)

def step_space(s: str) -> str:
    return re_space.sub(' ', s).strip()

# 2) 세 가지 순서 정의 ----------------------------------------------------------
def clean_order_1(s):
    # 기존 순서: 특수문자 제거 -> HTML 제거 -> 공백 정리
    if not isinstance(s, str): return ''
    s = step_special(s)
    s = step_html(s)
    s = step_space(s)
    return s

def clean_order_2(s):
    # 권장 순서: HTML 제거 -> 특수문자 제거 -> 공백 정리
    if not isinstance(s, str): return ''
    s = step_html(s)
    s = step_special(s)
    s = step_space(s)
    return s

def clean_order_3(s):
    # 변형: 공백 정리 -> 특수문자 제거 -> HTML 제거 (비권장 예)
    if not isinstance(s, str): return ''
    s = step_space(s)
    s = step_special(s)
    s = step_html(s)
    s = step_space(s)
    return s

# 3) 예시 문장으로 비교 ---------------------------------------------------------
example = """[속보]   <b>정부 발표</b>… '세금↑'?!   자세히 보기 👉 <a href='https://news.example.com/123?ref=home'>클릭</a>  <!--주석-->
<script>alert('x')</script>     #특집 2025-08-18 😀"""

ex_o1 = clean_order_1(example)
ex_o2 = clean_order_2(example)
ex_o3 = clean_order_3(example)

print("=== 예시 문장 원본 ===\n", example, "\n")
print("[Order1] 특수문자→HTML→공백:", ex_o1)
print("[Order2] HTML→특수문자→공백:", ex_o2)
print("[Order3] 공백→특수문자→HTML:", ex_o3)
```
