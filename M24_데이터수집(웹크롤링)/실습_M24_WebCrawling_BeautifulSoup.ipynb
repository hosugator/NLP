## ì˜¤ëŠ˜ì˜ ì¦ê¶Œë‰´ìŠ¤ ë°ì´í„° ìˆ˜ì§‘

```python
# 1.íŒ¨í‚¤ì§€ ê°€ì ¸ì˜¤ê¸°
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# 2. URLì„ ì—´ì–´ BeautifulSoupë¡œ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°

# ë„¤ì´ë²„ ì¦ê¶Œ ì‚¬ì´íŠ¸
url = 'https://finance.naver.com/'

# response = urlopen('URL ì£¼ì†Œ ì…ë ¥')
response = urlopen(url)

soup = BeautifulSoup(response, 'html.parser') #html.parser, lxml, lxml-xml, html5lib
print(soup.prettify())
```

```python
# 3-1. ì›í•˜ëŠ” ë‰´ìŠ¤ ì²«ë²ˆì§¸ ë‚´ìš© ì°¾ì•„ ì¶œë ¥í•˜ê¸°

# 'section_strategy' í´ë˜ìŠ¤ë¥¼ ê°€ì§„ ìš”ì†Œ ì°¾ê¸°
section_strategy = soup.find('div', class_='section_strategy')
print(section_strategy)
# 'section_strategy' í´ë˜ìŠ¤ ë‚´ì— a tagë¥¼ ê°€ì§„ ìš”ì†Œ ì°¾ê¸° 
a_tag = section_strategy.find_next('a') 
data = a_tag.get_text(strip=True)
print(data)
```

```python
# 3-2. ì›í•˜ëŠ” ë‰´ìŠ¤ ëª©ë¡ ëª¨ë‘ ì°¾ì•„ ì¶œë ¥í•˜ê¸°

# 'section_strategy' í´ë˜ìŠ¤ë¥¼ ê°€ì§„ ìš”ì†Œ ì°¾ê¸°
section_strategy = soup.find('div', class_='section_strategy')

# h2 íƒœê·¸ í•˜ìœ„ì˜ ëª¨ë“  'a' íƒœê·¸ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
a_tags = section_strategy.find_all('a')
print(a_tags)
# ê° 'a' íƒœê·¸ì˜ í…ìŠ¤íŠ¸ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
text =[]
for a_tag in a_tags:
    print(a_tag.get_text(strip=True))
    text.append(a_tag.get_text(strip=True))
```

```python
# 4. ì €ì¥í•  íŒŒì¼ ìƒì„± stock_news.txt íŒŒì¼ ìƒì„±
with open("stock_news.txt", 'w') as f:
    f.write("ì˜¤ëŠ˜ì˜ ì¦ì‹œ ë‰´ìŠ¤ ë°ì´í„°\n")
    f.writelines(text)
```
